```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error, r2_score
```

```python
add_data = pd.read_csv("reklama_sales.csv")

add_df = pd.DataFrame(add_data)

add_df
```

```python
tv = add_df[['TV']]
radio = add_df[['Radio']]
newspaper = add_df[['Newspaper']]
sales = add_df['Sales']
plt.scatter(tv, sales, color="blue")
plt.scatter(radio, sales, color="red")
plt.scatter(newspaper, sales, color="green")

plt.title("Commercial Prises")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
tv_x_train, tv_x_test, tv_y_train, tv_y_test = train_test_split(tv, sales, test_size=0.2, random_state=42)

tv_x_train, tv_x_test

```

```python
radio_x_train, radio_x_test, radio_y_train, radio_y_test= train_test_split(radio, sales, test_size=0.2, random_state=42)

radio_x_train, radio_x_test
```

```python
bilb_x_train, bilb_x_test, bilb_y_train, bilb_y_test= train_test_split(newspaper, sales, test_size=0.2, random_state=42)

bilb_x_train,bilb_x_test
```

```python
bek_tv = LinearRegression()
bek_rad = LinearRegression()
bek_bilb = LinearRegression()

bek_tv.fit(tv_x_train,tv_y_train)
bek_rad.fit(radio_x_train, radio_y_train)
bek_bilb.fit(bilb_x_train, bilb_y_train)
```

```python
print("intercept:tv:", bek_tv.intercept_)
print("Slope:tv:", bek_tv.coef_[0])
print("intercept:rad:", bek_rad.intercept_)
print("Slope:tv:", bek_rad.coef_[0])
print("intercept:bilb:", bek_bilb.intercept_)
print("Slope:tv:", bek_bilb.coef_[0])
```

```python
tv_y_pred = bek_tv.predict(tv_x_test)

result_tv = pd.DataFrame({"actual":tv_y_test, "Predict": tv_y_pred})
result_tv
```

```python
rad_y_pred = bek_rad.predict(radio_x_test)

result_rad = pd.DataFrame({"actual":radio_y_test, "Predict": rad_y_pred})
result_rad
```

```python
bilb_y_pred = bek_bilb.predict(bilb_x_test)

result_bilb = pd.DataFrame({"actual":bilb_y_test, "Predict": bilb_y_pred})
result_bilb
```

```python
mse_TV = mean_squared_error(tv_y_test, tv_y_pred)
r2_TV = r2_score(tv_y_test, tv_y_pred)
mse_RAD = mean_squared_error(radio_y_test, rad_y_pred)
r2_RAD = r2_score(radio_y_test, rad_y_pred)
mse_BILB = mean_squared_error(bilb_y_test, bilb_y_pred)
r2_BILB = r2_score(bilb_y_test, bilb_y_pred)
```

```python
print("Mean Squared Error:TV:", mse_TV)
print("R^2 Score:TV:", r2_TV)
print("Mean Squared Error:RADIO:", mse_RAD)
print("R^2 Score:RADIO:", r2_RAD)
print("Mean Squared Error:BILB:", mse_BILB)
print("R^2 Score:BILB:", r2_BILB)
```

```python
plt.scatter(tv_x_train, tv_y_train, color="blue", label="Taining Data TV")
plt.scatter(tv_x_test, tv_y_test, color="green", label="Tesing data TV")
plt.plot(tv, bek_tv.predict(tv), color="red", label="Regression Line TV")


plt.title("TV, RADIO AND COMMERCIALS LINEAR REGRESSION")
plt.xlabel("tv, radio, com")
plt.ylabel("sales")
plt.legend()

plt.show
```

```python
plt.scatter(radio_x_train, radio_y_train, color="yellow", label="Taining Data RADIO")
plt.scatter(radio_x_test, radio_y_test, color="gray", label="Tesing data RADIO")
plt.plot(radio, bek_rad.predict(radio), color="blue", label="Regression Line RADIO",linewidth = '5')
plt.title("TV, RADIO AND COMMERCIALS LINEAR REGRESSION")
plt.xlabel("tv, radio, com")
plt.ylabel("sales")
plt.legend()

plt.show
```

```python
plt.scatter(bilb_x_train, bilb_y_train, color="red", label="Taining Data BILBOADRD")
plt.scatter(bilb_x_test, bilb_y_test, color="black", label="Tesing data BILBOARD")
plt.plot(newspaper, bek_bilb.predict(newspaper), color="Orange", label="Regression Line BILBOARD")
plt.title("TV, RADIO AND COMMERCIALS LINEAR REGRESSION")
plt.xlabel("tv, radio, com")
plt.ylabel("sales")
plt.legend()

plt.show
```
